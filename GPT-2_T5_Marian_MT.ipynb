```
model = AutoModelWithLMHead.from_pretrained("t5-base") #loading t5 model in translation mode eng->de
translator = pipeline("translation_en_to_de")

model_name = 'Helsinki-NLP/opus-mt-de-pl'  #loading marian  model translator de->pl
tokenizer = MarianTokenizer.from_pretrained(model_name)
model_m = MarianMTModel.from_pretrained(model_name)
```

# 4. Initiating auxilary funcions

```
def temp_list(text_out): # creating format to write in text file
  text_all=[]
  next_line=''
  dot_line='...................................'
  text_all.append(next_line)
  text_all.append(dot_line)
  for t in text_out:
    text_all.append(t)
  return text_all

temp_de =[]
file_name= '/content/drive/MyDrive/GPT/en_de_pl_trans.txt'
def save_Table_to_file(file_name, temp_file=temp_de):
    with open(file_name, 'a') as f:
        for item in temp_file:
            f.write("%s\n" % item)

def translate_ge_to_pl(src_text):
    translated = model_m.generate(**tokenizer.prepare_seq2seq_batch(src_text, return_tensors="pt"))
    tgt_text = [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
    return tgt_text

def trans_en_de(text_out): #translating english to germant with the use of T5
  temp=[]
  for text in text_out:
    text_out_de= translator(text)

    src_text= text_out_de[0]['translation_text']
    temp.append(src_text)
  return temp
```

# 5. Generating and translating text

```
#@title Generating and translating give initial english text { display-mode: "form" }
give_run_name = 'run355' #@param ["None"] {type:"raw", allow-input: true}
paste_prefix_text = "Now you need to close your eyes and" #@param [""] {allow-input: true}
set_temperature = 0.5 #@param {type:"slider", min:0.1, max:1, step:0.05}
set_text_length = 420 #@param {type:"slider", min:0, max:1023, step:10}
set_nr_samples = 5 #@param {type:"slider", min:4, max:10, step:1}



print(" 1. ==================>>> please wait text is generated..." ) 

text_out=gpt2.generate(sess,
              run_name=give_run_name,
              length=set_text_length,
              temperature=set_temperature,
              prefix=paste_prefix_text,
              nsamples=set_nr_samples,
              batch_size=set_nr_samples,
              return_as_list=True
              )

for t in text_out:
  print(t)
print("\n 2.  ==================>>> Text is being translated to German...")
text_out_de= trans_en_de(text_out)
for t in text_out_de:
  print(t)
print("\n 3.  ==================>>> Text is being translated to Polish...")
text_out_pl= translate_ge_to_pl(text_out_de)
for t in text_out_pl:
  print(t)
print("\n . ==================>>> If you like this seesion you could save that to file... <<<<==================")
```

```
#@title Save generated and translated text to file given below { display-mode: "form" }

text_file = "/content/drive/MyDrive/GPT/en_de_pl_trans.txt" #@param [""] {allow-input: true}

text_file= file_name



text_temp = temp_list(text_out)
save_Table_to_file(file_name, temp_file=text_temp) #saving eng

text_temp = temp_list(text_out_de)
save_Table_to_file(file_name, temp_file=text_temp) #saving ge

text_temp = temp_list(text_out_pl)
save_Table_to_file(file_name, temp_file=text_temp) #saving pl
print('Generated and translated text were saved to file {}'.format(file_name))
```
