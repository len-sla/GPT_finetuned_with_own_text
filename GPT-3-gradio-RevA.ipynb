## testing response

```
def gpt_output(prompt):
    # prompt="First response from model"
    response= openai.Completion.create(
        model="text-davinci-003",
        prompt=prompt,
        temperature=0.9,
        max_tokens=150,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0.6,
        stop=[" Human :", " AI:"]

        )
    print("\n --- Model response: ---",response.choices[0].text)
    return response.choices[0].text


```

```
prompt="Tell me a joke about AI"
gpt_output(prompt)
```

```
while True:
    query = input("Ask me a question:\n\n ")
    gpt_output(query)
```

# 2. Gradio-front-end with GPT3 text  chatboot

```
def gpt_output_1(prompt):
    # prompt="First response from model"
    response= openai.Completion.create(
        model="text-davinci-003",
        prompt=prompt,
        temperature=0.9,
        max_tokens=150,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0.6,
        stop=[" Human :", " Model:"]

        )
        # print("\n --- Model response: ---",response.choices[0].text)
    return response.choices[0].text
```

```
def chat_gpt_clone(input, history):
    history = history or []
    s= list(sum(history, ()))
    s.append(input)
    inp = ''.join(s)
    output = gpt_output_1(inp)
    history.append((input, output))
    return history,history
```

```
 block = gr.Blocks()
 with block:
     
     gr.Markdown("""<h1><center> --- That is your virtual assistent --- </center></h1>""")
     chatbot = gr.Chatbot()
     message  = gr.Textbox(placeholder=prompt)
     state    = gr.State()
     submit   = gr.Button("SEND")

     submit.click(chat_gpt_clone, [message, state], [chatbot, state])
block.launch(debug=True, share=True) # share=True creating public link
```
